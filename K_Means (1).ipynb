```python
X.columns
```

```python
X_new = pd.DataFrame()

for name in X.columns :
  # print(name)

  # ë°ì´í„°ê°€ ë¬¸ìžì—´ì´ë©´, ë°ì´í„°ì˜ ì¢…ë¥˜ê°€ ëª‡ê°œì¸ì§€ í™•ì¸í•œë‹¤.
  if X[ name ].dtype == object :

    if X[name].nunique() >= 3 :
      # ì›í•« ì¸ì½”ë”©í•œë‹¤.
      ct = ColumnTransformer( [('encoder', OneHotEncoder() ,[0] )] , 
                       remainder= 'passthrough' )
      
      col_names = sorted( X[name].unique() )

      X_new[col_names] = ct.fit_transform(  X[name].to_frame()  )

    else :
      # ë ˆì´ë¸” ì¸ì½”ë”© í•œë‹¤.
      label_encoder = LabelEncoder()
      X_new[name] = label_encoder.fit_transform( X[name] )
  
  # ìˆ«ìž ë°ì´í„°ì¼ë•Œì˜ ì²˜ë¦¬ëŠ” ì—¬ê¸°ì„œ
  else :
    X_new[name] = X[name]
```
